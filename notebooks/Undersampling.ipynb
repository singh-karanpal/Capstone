```python
import numpy as np
import pandas as pd
```

```python
X_train_Q1 = pd.read_excel('../data/interim/X_train_Q1_clean.xlsx')
X_valid_Q1 = pd.read_excel('../data/interim/X_valid_Q1_clean.xlsx')

y_train_Q1 = pd.read_excel('../data/interim/y_train_Q1.xlsx')
y_valid_Q1 = pd.read_excel('../data/interim/y_valid_Q1.xlsx')
```

```python
df = pd.concat([X_train_Q1, y_train_Q1.iloc[:,0:12]], axis = 1)
```

```python
df.shape
```

```python
labels = []
records = []

for i in df.columns[1:]:
    labels.append(i)
    records.append(np.sum(df[i]))
```

```python
distribution = pd.DataFrame({'labels':labels, 'records': records})
distribution.sort_values(by = 'records', ascending = False)
```

### Let's build a dataset with 500 records for each Label

```python
num_records = 300 #to get 500 records per label
```

```python
sub_df = pd.DataFrame()

for i in df.columns[1:]:
    temp = df.query(f'{i} == 1').sample(num_records)
    sub_df = pd.concat([sub_df, temp])
```

### Removing these duplicates

```python
sub_df = sub_df.drop_duplicates(keep = 'first')
```

```python
labels = []
records = []

for i in sub_df.columns[1:]:
    labels.append(i)
    records.append(np.sum(sub_df[i]))
    
distribution = pd.DataFrame({'labels':labels, 'records': records})
distribution.sort_values(by = 'records', ascending = False)
```
